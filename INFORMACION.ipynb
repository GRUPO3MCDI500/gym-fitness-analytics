## 2. Instalación de Anaconda

Anaconda facilita la instalación de Python y de herramientas comunes para ciencia de datos.

### Pasos de instalación

1. Abrir el navegador web.
2. Buscar en Google: `Anaconda Download`.
3. Ingresar al sitio oficial de Anaconda.
4. Descargar la versión para Windows.
5. Ejecutar el archivo descargado.
6. Presionar `Next`.
7. Aceptar los términos de instalación.
8. Seleccionar la opción `Just Me`.
9. Mantener la carpeta de instalación sugerida por defecto.
10. Dejar marcada la opción para registrar Anaconda como Python principal.
11. Finalizar la instalación.

Una vez instalado, se debe abrir el programa llamado `Anaconda Navigator`.

---

## 3. Abrir Jupyter Notebook desde Anaconda

Para ejecutar el notebook del proyecto se debe utilizar Jupyter Notebook desde Anaconda.

### Pasos

1. Abrir el menú de inicio de Windows.
2. Buscar `Anaconda Navigator`.
3. Abrir Anaconda Navigator.
4. Buscar la opción `Jupyter Notebook`.
5. Presionar `Launch`.

Esto abrirá Jupyter Notebook en el navegador web. Aunque se vea en el navegador, estará usando Python instalado localmente en el computador.
