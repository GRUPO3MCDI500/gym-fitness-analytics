# 4. Relación con el foro técnico

El foro técnico de Fase 3 pide discutir criterios para decidir entre programación estructurada y recursiva.

En el proyecto, la conclusión técnica es:

- Para recorridos simples sobre datos tabulares, conviene programación estructurada o Pandas.
- Para problemas divisibles en subproblemas, la recursividad puede ser útil.
- La recursividad no siempre es más eficiente.
- Debe analizarse tiempo y memoria.
- La legibilidad es importante en proyectos colaborativos.

Estas ideas deben mencionarse en el informe.
