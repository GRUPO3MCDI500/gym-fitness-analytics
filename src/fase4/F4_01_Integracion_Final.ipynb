# F4_01_Integracion_Final

Notebook de integración final para la Evaluación Sumativa Fase 4.

Este notebook conecta los avances F1-F4, valida el dataset procesado y genera tablas de trazabilidad que sirven como evidencia para el informe final, el README y la presentación audiovisual.

**Resultado esperado:** dataset cargado, columnas validadas, resumen general y tablas de integración listas para documentar.

```python
from pathlib import Path
import pandas as pd
import numpy as np

def encontrar_raiz_proyecto(nombre_repo="gym-fitness-analytics"):
    ruta_actual = Path.cwd().resolve()
    for ruta in [ruta_actual] + list(ruta_actual.parents):
        if ruta.name == nombre_repo:
            return ruta
    raise FileNotFoundError(f"No se pudo encontrar la raíz del proyecto '{nombre_repo}' desde {ruta_actual}")

PROJECT_ROOT = encontrar_raiz_proyecto()
DATA_PATH = PROJECT_ROOT / "data" / "processed" / "gym_data_processed.csv"

if not DATA_PATH.exists():
    raise FileNotFoundError(f"No se encontró el dataset procesado en la ruta: {DATA_PATH}")

df = pd.read_csv(DATA_PATH)
print("Dataset cargado correctamente.")
print("Filas:", df.shape[0])
print("Columnas:", df.shape[1])
df.head()

REPORTS_DIR = PROJECT_ROOT / "reports" / "fase4"
FIGURES_DIR = REPORTS_DIR / "figures"
EVIDENCIAS_DIR = REPORTS_DIR / "evidencias"
FIGURES_DIR.mkdir(parents=True, exist_ok=True)
EVIDENCIAS_DIR.mkdir(parents=True, exist_ok=True)

print("Raíz del proyecto:", PROJECT_ROOT)
print("Reportes F4:", REPORTS_DIR)
```

## Validación de columnas necesarias para F4

```python
columnas_f4 = [
    "country", "region", "year", "gym_memberships",
    "fitness_participation_rate", "total_health_club_revenue_usd",
    "number_of_gyms", "gym_penetration_rate", "gdp_per_capita_usd",
    "population_total", "average_membership_cost_usd",
    "memberships_per_100k", "gyms_per_100k",
    "revenue_per_membership_usd", "periodo"
]
faltantes = [col for col in columnas_f4 if col not in df.columns]
if faltantes:
    raise ValueError(f"Faltan columnas necesarias para F4: {faltantes}")
print("Todas las columnas necesarias para F4 están disponibles.")
```
