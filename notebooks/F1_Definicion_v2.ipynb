## Visualizaciones preliminares

Las visualizaciones iniciales permiten observar patrones generales. En esta fase deben entenderse como exploratorias, ya que todavía no se ha realizado limpieza completa del dataset.

### Evolución global de membresías de gimnasio

Este gráfico permite observar la evolución agregada de membresías por año.

````python
if {"year", "gym_memberships"}.issubset(df.columns):
    memberships_by_year = df.groupby("year", as_index=False)["gym_memberships"].sum()

    plt.figure(figsize=(10, 5))
    plt.plot(memberships_by_year["year"], memberships_by_year["gym_memberships"], marker="o")
    plt.title("Evolución global de membresías de gimnasio")
    plt.xlabel("Año")
    plt.ylabel("Membresías de gimnasio")
    plt.grid(True, alpha=0.3)
    plt.show()
else:
    print("No se puede generar el gráfico porque faltan columnas requeridas.")
````

#### Interpretación preliminar

La visualización permite observar la tendencia general de las membresías en el periodo analizado. Sin embargo, antes de formular conclusiones será necesario revisar valores atípicos, consistencia temporal y cobertura de países por año.

### Penetración promedio de gimnasios por región

Este gráfico permite comparar de forma preliminar la tasa promedio de penetración de gimnasios entre regiones.

````python
if {"region", "gym_penetration_rate"}.issubset(df.columns):
    penetration_by_region = (
        df.groupby("region", as_index=False)["gym_penetration_rate"]
        .mean()
        .sort_values("gym_penetration_rate", ascending=False)
    )

    plt.figure(figsize=(10, 5))
    plt.bar(penetration_by_region["region"], penetration_by_region["gym_penetration_rate"])
    plt.title("Penetración promedio de gimnasios por región")
    plt.xlabel("Región")
    plt.ylabel("Tasa promedio de penetración")
    plt.xticks(rotation=45, ha="right")
    plt.tight_layout()
    plt.show()

    display(penetration_by_region)
else:
    print("No se puede generar el gráfico porque faltan columnas requeridas.")
````

#### Interpretación preliminar

La comparación regional permite identificar diferencias iniciales entre zonas geográficas. Estas diferencias deberán analizarse considerando factores económicos, población, urbanización y disponibilidad de registros por región.

### Relación preliminar entre PIB per cápita y penetración de gimnasios

#### Interpretación preliminar
