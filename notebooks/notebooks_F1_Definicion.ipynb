```python
# Carga de librerías principales

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from pathlib import Path

print("Librerías cargadas correctamente.")
```

```python
from pathlib import Path

# Ruta relativa desde la carpeta notebooks/ hacia data/raw/
DATA_PATH = Path("../data/raw/clean_gym_data.csv")

if DATA_PATH.exists():
    print(f"Archivo encontrado: {DATA_PATH}")
else:
    print("No se encontró el archivo. Verifica que el dataset esté en data/raw/clean_gym_data.csv")
```

```python
# Carga del dataset

df = pd.read_csv(DATA_PATH)

print("Dataset cargado correctamente.")
```

```python
# Dimensiones iniciales del dataset

filas, columnas = df.shape

print(f"Cantidad de filas: {filas}")
print(f"Cantidad de columnas: {columnas}")
```

```python
# Visualización de las primeras filas

df.head()
```

```python
# Revisión de nombres de columnas

columnas_dataset = df.columns.tolist()

for columna in columnas_dataset:
    print(columna)
```

```python
# Información general del dataset

df.info()
```

```python
# Revisión de valores nulos por columna

valores_nulos = df.isnull().sum().sort_values(ascending=False)

valores_nulos
```

```python
# Resumen estadístico de variables numéricas

df.describe()
```

```python
# Revisión de duplicados

duplicados = df.duplicated().sum()

print(f"Cantidad de filas duplicadas: {duplicados}")
```

```python
# Función simple de exploración inicial

def resumen_inicial(dataframe):
    """
    Genera un resumen inicial del dataset.
    
    Retorna:
    - cantidad de filas
    - cantidad de columnas
    - cantidad total de valores nulos
    - cantidad de columnas numéricas
    - cantidad de columnas categóricas
    - cantidad de filas duplicadas
    """
    
    resumen = {
        "filas": dataframe.shape[0],
        "columnas": dataframe.shape[1],
        "valores_nulos_totales": int(dataframe.isnull().sum().sum()),
        "columnas_numericas": dataframe.select_dtypes(include="number").shape[1],
        "columnas_categoricas": dataframe.select_dtypes(exclude="number").shape[1],
        "filas_duplicadas": int(dataframe.duplicated().sum())
    }
    
    return resumen

resumen_inicial(df)
```

```python
# Revisión del rango temporal disponible

if "year" in df.columns:
    print("Año mínimo:", df["year"].min())
    print("Año máximo:", df["year"].max())
    print("Cantidad de años registrados:", df["year"].nunique())
else:
    print("La columna 'year' no se encuentra en el dataset.")
```

```python
# Revisión de países y regiones disponibles

if "country" in df.columns:
    print("Cantidad de países:", df["country"].nunique())
else:
    print("La columna 'country' no se encuentra en el dataset.")

if "region" in df.columns:
    print("\nRegiones disponibles:")
    for region in df["region"].dropna().unique():
        print("-", region)
else:
    print("La columna 'region' no se encuentra en el dataset.")
```

```python
# Identificación de variables principales esperadas

variables_esperadas = [
    "country",
    "region",
    "year",
    "gym_memberships",
    "fitness_participation_rate",
    "total_health_club_revenue_usd",
    "number_of_gyms",
    "gym_penetration_rate",
    "urban_population_percentage",
    "obesity_rate",
    "gdp_per_capita_usd",
    "population_total",
    "average_membership_cost_usd",
    "insufficient_physical_activity_pct"
]

variables_presentes = [col for col in variables_esperadas if col in df.columns]
variables_faltantes = [col for col in variables_esperadas if col not in df.columns]

print("Variables esperadas presentes:")
for col in variables_presentes:
    print("-", col)

print("\nVariables esperadas no encontradas:")
for col in variables_faltantes:
    print("-", col)
```

```python
# Primer análisis descriptivo: membresías globales por año

if "year" in df.columns and "gym_memberships" in df.columns:
    membresias_por_anio = df.groupby("year")["gym_memberships"].sum().reset_index()
    membresias_por_anio.head()
else:
    print("No se encontraron las columnas necesarias: 'year' y/o 'gym_memberships'.")
```

```python
# Visualización inicial: evolución global de membresías

if "year" in df.columns and "gym_memberships" in df.columns:
    plt.figure(figsize=(10, 5))
    plt.plot(membresias_por_anio["year"], membresias_por_anio["gym_memberships"], marker="o")
    plt.title("Evolución global de membresías de gimnasio")
    plt.xlabel("Año")
    plt.ylabel("Membresías de gimnasio")
    plt.grid(True)
    plt.show()
else:
    print("No se puede generar el gráfico porque faltan columnas necesarias.")
```

```python
# Primer análisis descriptivo: ingresos globales por año

columna_ingresos = "total_health_club_revenue_usd"

if "year" in df.columns and columna_ingresos in df.columns:
    ingresos_por_anio = df.groupby("year")[columna_ingresos].sum().reset_index()
    ingresos_por_anio.head()
else:
    print(f"No se encontraron las columnas necesarias: 'year' y/o '{columna_ingresos}'.")
```

```python
# Visualización inicial: evolución global de ingresos del sector

if "year" in df.columns and columna_ingresos in df.columns:
    plt.figure(figsize=(10, 5))
    plt.plot(ingresos_por_anio["year"], ingresos_por_anio[columna_ingresos], marker="o")
    plt.title("Evolución global de ingresos de la industria fitness")
    plt.xlabel("Año")
    plt.ylabel("Ingresos totales en USD")
    plt.grid(True)
    plt.show()
else:
    print("No se puede generar el gráfico porque faltan columnas necesarias.")
```

```python
# Comparación inicial: promedio de penetración de gimnasios por región

if "region" in df.columns and "gym_penetration_rate" in df.columns:
    penetracion_region = (
        df.groupby("region")["gym_penetration_rate"]
        .mean()
        .sort_values(ascending=False)
        .reset_index()
    )
    
    penetracion_region
else:
    print("No se encontraron las columnas necesarias: 'region' y/o 'gym_penetration_rate'.")
```

```python
# Visualización inicial: promedio de penetración de gimnasios por región

if "region" in df.columns and "gym_penetration_rate" in df.columns:
    plt.figure(figsize=(10, 5))
    plt.bar(penetracion_region["region"], penetracion_region["gym_penetration_rate"])
    plt.title("Promedio de penetración de gimnasios por región")
    plt.xlabel("Región")
    plt.ylabel("Tasa promedio de penetración")
    plt.xticks(rotation=45)
    plt.grid(axis="y")
    plt.show()
else:
    print("No se puede generar el gráfico porque faltan columnas necesarias.")
```
